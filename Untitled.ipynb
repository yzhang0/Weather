```python
import pandas as pd

def createCityRow(dataset, city):
    fname = city + ".csv"
    raw_data = pd.read_csv(fname, header=None)
    raw_data = raw_data[[0, 2]]
    df = pd.DataFrame(columns=['date', city])
    date = ""
    index = 0
    for i, r in raw_data.iterrows():
        if not raw_data.loc[i][0].isdigit():
            2011-01-01
            mm = raw_data.loc[i][0]
        elif raw_data.loc[i][0].isdigit() and raw_data.loc[i][0] != "2017":
            df.loc[index] = [str(raw_data.loc[i][0]) + " " + mm, raw_data.loc[i][2]]
            index += 1
    if dataset.empty:
        return df
    else:
        dataset = pd.merge(dataset, df, on="date", how="left")
        return dataset

def changeMapping(df):
    df = df.set_index('date')
    df = pd.DataFrame(df.stack().reset_index())
    df.columns = ['date', 'city', 'fahrenheit']
    df = df[['city', 'date', 'fahrenheit']]
    return df

def main():
    cities = sorted([
                     'Montreal',
                     'NewYork',
                     'SanFrancisco',
                     'Toronto',
                     'Vancouver',
                     'Shanghai',
                     'Beijing',
                     'NewDelhi',
                     'Tokyo',
                     'RiodeJaneiro',
                     'Moscow',
                     'Seattle',
                     'London',
                     'Bangkok',
                     'Paris',
                     'Singapore',
                     'Sydney',
                     'Boston',
                     'Madrid',
                     'BuenosAires'])
    df = pd.DataFrame()
    for city in cities:
        print city
        df = createCityRow(df, city)
    df = changeMapping(df)
    df.to_csv('All.csv', index=False)

main()
```
